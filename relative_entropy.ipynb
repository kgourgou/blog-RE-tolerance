# A tolerance for the relative entropy (for univariate gaussian distributions)

The following is for a short demonstration I do [here](http://notabstractnonsense.com/math,%20research/2015/08/03/a-tolerance-for-the-relative-entropy.html).

```python
mean = "mean"
std  = "std" # Standard deviation.


def gaussian_re(p,q):
    # Calculates the relative entropy between two univariate gaussians
    result = np.log(q[std]/p[std])+(p[std]**2 + (p[mean]-q[mean])**2)/(2*q[std]**2)-0.5
    return result

def pdf(x,p):
    # Calculates the gaussian probability density at x
    result = np.exp(-(x-p[mean])**2/(2*p[std]**2))/(np.sqrt(2*np.pi*p[std]))
    return result
```

```python
p = {mean : 1, std : 1e+4}
q = {mean : 10, std : 1e+4}

print 'R(q|p)=',gaussian_re(p,q)

x = np.linspace(-40,40);
pl.plot(x,pdf(x,p))
pl.plot(x,pdf(x,q))
f = pl.gca()

f.axes.get_yaxis().set_ticklabels([]);
```

```python
p = {mean : 0, std : 0.31};
q = {mean : 0, std : 0.31};

print 'R(p|q)=',gaussian_re(p,q)

x = np.linspace(-3,3);
pl.plot(x,pdf(x,p))
pl.plot(x,pdf(x,q))

f = pl.gca();

f.axes.get_yaxis().set_ticklabels([])
```
